```python
import re
import json
```

```python
def convert_to_jsonl(input_file, output_file):
    # Pattern to match: starting digits, a dot or parenthesis, and then the question text
    # It accounts for potential multiple lines by using re.DOTALL if needed, 
    # but here we assume one question per paragraph.
    question_pattern = re.compile(r'^\d+[\.\)]\s*(.*)', re.MULTILINE)
    
    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()

    # Find all matches for the questions
    questions = question_pattern.findall(content)

    with open(output_file, 'w', encoding='utf-8') as f:
        for q in questions:
            # Clean up extra whitespace or newlines from each question
            clean_q = q.strip()
            if clean_q:
                # Create a dictionary for each line
                json_record = {"question": clean_q}
                # Write as a single line in the JSONL file
                f.write(json.dumps(json_record) + '\n')

# Usage
convert_to_jsonl('questions.txt', 'questions.jsonl')
```
